```python
# [SPATIOTEMPORAL] SPL4SMGP.007 SMAP L4 Global 3-hourly 9-km Surface and Root Zone Soil Moisture
smap_product = 'NASA/SMAP/SPL4SMGP/008'
smap_bands = ['sm_surface', 'sm_rootzone', 'sm_profile', 'sm_surface_wetness', 'sm_rootzone_wetness', 'sm_profile_wetness', 'surface_temp',
          'soil_temp_layer1', 'soil_temp_layer2', 'soil_temp_layer3', 'soil_temp_layer4', 'soil_temp_layer5', 'soil_temp_layer6']

```

```python
# show sm_surface from smap_product as a layer on an interactive map
smap = ee.ImageCollection(smap_product).select(smap_bands)
smap_surface = smap.select('sm_surface')
smap_surface
```

```python
import geemap

# Create an interactive map
Map = geemap.Map()

# Add the SMAP surface soil moisture image collection to the map.
# I will select a single image from the collection for visualization.
# You can adjust the date range to select a different image.
first_image = smap_surface.first()

# Define visualization parameters (adjust as needed)
vis_params = {
    'min': 0,
    'max': 0.5,
    'palette': ['0502ff', '33ccff', '05ff19', 'fffb1a', 'ff0c05']
}

# Add the image layer to the map
Map.addLayer(first_image, vis_params, 'SMAP Surface Soil Moisture')

# Display the map
Map
```
